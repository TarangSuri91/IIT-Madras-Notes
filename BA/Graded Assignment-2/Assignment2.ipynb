**Primal**

```python
import pulp as p
```

```python
primal = p.LpProblem('Primal',p.LpMaximize)
```

```python
A = p.LpVariable("A",lowBound=0)
B = p.LpVariable("B",lowBound=0)
```

```python
primal += 150*A + 200*B
```

```python
primal += 3*A + 5*B <= 22000
primal += 2*A + 3*B <= 15000
primal += 3*A + 2*B <= 20000
```

```python
print(primal)
```

```python
solution = primal.solve()
print(p.LpStatus[solution])
```

```python
print(p.value(A),p.value(B),p.value(primal.objective))
```

```python
150 * p.value(A) + 200 * p.value(B)
```

### Dual

```python
dual = p.LpProblem('Dual',p.LpMinimize)
```

```python
y1 = p.LpVariable("y1",lowBound=0)
y2 = p.LpVariable("y2",lowBound=0)
y3 = p.LpVariable("y3",lowBound=0)
```

```python
dual += 1000 * (22 * y1 + 15 * y2 + 20 * y3)
```

```python
dual += 3*y1 + 2*y2 + 3*y3 >= 150
dual += 5*y1 + 3*y2 + 2*y3 >= 200
```

```python
solution = dual.solve()
print(p.LpStatus[solution])
```

```python
print(p.value(y1),p.value(y2),p.value(y3), p.value(dual.objective))
```
