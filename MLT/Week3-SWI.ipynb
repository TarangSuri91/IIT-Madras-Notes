```python
import numpy as np
import pandas as pd
# from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
def nl_data(n):
    x = np.linspace(0,10,n)
    y_hat = 3 - 2 * x + x ** 2 #without noise
    y = (y_hat + 10 * np.random.rand(n)) #with noise
    return x, y, y_hat
```

```python
x, y, y_hat = nl_data(10)
x.shape, y.shape
```

```python
# f = plt.figure(figsize=(4,4))
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("y")

plt.plot(x, y_hat, c='r')
```

```python
def add_dummy(X):
    return np.column_stack((np.ones(X.shape[0]), X))
```

```python
X = np.array([[1,2], [3,4]])
print(X.shape)
add_dummy(X)
```

```python
def fit(X, y):
    print(X.shape)
    return np.linalg.pinv(X) @ y
```

```python
x = add_dummy(x)
w = fit(x, y)
print(w)
```

```python
def predict(X, w):
    print(X.shape, w.shape)
    return X @ w
```

```python
y_predict = predict(x, w)
```

```python
x, y, y_predict
```

```python
plt.scatter(x[:,-1], y)
plt.plot(x[:,-1], y_predict, c='r')
```
