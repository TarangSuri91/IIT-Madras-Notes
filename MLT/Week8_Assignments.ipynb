```python
def hinge_loss(y_test, y_pred):
    return np.mean([max(0, 1 - x * y) for x, y in zip(y_test, y_pred)])
```

```python
import numpy as np
def solve_eqn(A, b):
    A = np.column_stack((A, np.ones(A.shape[0])))
   
    S = [[i @ j for j in A] for i in A]

    x = np.linalg.solve(S, b)

    w = np.zeros_like(x)
    for i in range(len(x)):
        w += x[i] * A[i]
    return w
```

```python
A = np.array([[1, 0], [3, 1], [3, -1]])
b = np.array([-1, 1, 1])

print(solve_eqn(A, b))
```
