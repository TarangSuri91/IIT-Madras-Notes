```python
import numpy as np
```

```python
def euclid(a, b):
    return np.sum((a - b) ** 2, axis=1)

def manhattan(a, b):
    return np.sum(abs(a - b), axis=1)
```

```python
X = np.array([[5, 6],[5, 8],[3, 8], [2,4], [3,3], [4,2]]) #input points
y = np.array([1,1,1,2,2,2]) # input classes
new_example = np.array([2,6])
```

```python
points = dict(zip(euclid(X, new_example), y))
print("All points are ", points)

least_3 = sorted(points.items())[:3]
print(f"Points nearest to the new examples are {least_3}")

# To find the output label, find the class that contains the majority points.
# Thus if the list of points are [(4, 2), (5, 1), (9, 1)], majority class is 1.
```
