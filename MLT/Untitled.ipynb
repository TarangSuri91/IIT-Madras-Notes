```python
import numpy as np
def gini_index(dict):
    total = sum(dict.values())
    total_p = 0
    for each in dict.keys():
        p = dict.values[each]/total
        total_p += p * (1 - p)
        
    return total_p
```

```python
X = np.array([[1,2],[1,3]])
w = np.array([0.1,0.2])
y = np.array([1,0])
```

```python
def linear_combination(X, w):
    return X @ w
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def activation(X, w):
    return sigmoid(linear_combination(X, w))
```

```python
def calculate_gradient(X, y, w, reg_rate=0):
    return X.T @ (activation(X, w) - y) + reg_rate * w
```

```python
calculate_gradient(X, y, w)
```
