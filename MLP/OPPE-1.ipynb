```python
numeric_features = ['Adult_Mortality',	'infant_deaths',	'Alcohol',	'percentage_expenditure',	'Hepatitis_B',	'Measles',	'BMI',	'under-five_deaths', 'Polio',	'Total_expenditure',	'Diphtheria',	'HIV_AIDS',	'GDP',	'Population',	'Income_composition_of_resources',	'Schooling']
categorical_features = ['Country', 'Status']
```

```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

numeric_transfomer = Pipeline([('si', SimpleImputer(missing_values=np.nan, strategy='mean')), ('ss', StandardScaler())])
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

p = ColumnTransformer(transformers=[('num', numeric_transfomer, numeric_features), 
                                ('cat', categorical_transformer, categorical_features)], 
                  remainder='passthrough')
X_train = p.fit_transform(X_train)
X_test = p.transform(X_test)
```

Q.19

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error


sgd = SGDRegressor(penalty='l1', alpha=0.001, learning_rate='constant', eta0=0.001, random_state=42)

sgd.fit(X_train, y_train)
mse_train = mean_squared_error(y_train, sgd.predict(X_train))
mse_test = mean_squared_error(y_test, sgd.predict(X_test))

print(abs(mse_train - mse_test))
```

Q.20

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score

sgd = SGDRegressor(penalty='l2', alpha=0.01, learning_rate='adaptive', eta0=0.01, random_state=42, loss='huber')

sgd.fit(X_train, y_train)
r2_test = r2_score(y_test, sgd.predict(X_test))

print(r2_test)
```

Q.21

```python
from sklearn.model_selection import ShuffleSplit

sgd = SGDRegressor(penalty='l2', alpha=0.01, learning_rate='adaptive', eta0=0.01, random_state=42, loss='huber')

ss = ShuffleSplit(n_splits=10, test_size=0.2, random_state=42)

scores = cross_val_score(sgd, X_train, y_train, cv=ss)

np.sqrt(np.var(scores))
```
