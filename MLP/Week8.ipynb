```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pandas.plotting import scatter_matrix

from sklearn.preprocessing import OrdinalEncoder, StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, StratifiedShuffleSplit
from keras.datasets import mnist
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, ConfusionMatrixDisplay, classification_report, plot_roc_curve, roc_auc_score
from sklearn.pipeline import Pipeline
```

## Pre-processing data

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
fig = plt.figure(figsize=(10, 7))
for i in range(15):
  ax = fig.add_subplot(3, 5, i + 1)
  ax.imshow(X_train[i], cmap=plt.get_cmap('gray'))
  ax.set_title(f'Label y: {y_train[i]}')
  plt.axis('off')
```

```python
X_train.shape[0]
```

```python
X_train = X_train.reshape(X_train.shape[0], 28 * 28)
X_test = X_test.reshape(X_test.shape[0], 28 * 28)

#Normalizing
X_train = X_train/255
X_test = X_test/255
```

```python
X_train.shape, X_test.shape
```

```python
X_train = X_train[:10000, :]  # Consider the first 10K images in train
X_test = X_test[:2000, :]  # Consider the first 2K images in train
y_train = y_train[:10000]
y_test = y_test[:2000]
```

```python
X_train.shape, X_test.shape
```

## Linear SVM for MNIST multiclass classification

```python
pipe_1 = Pipeline([('scaler', MinMaxScaler()), 
                   ('classifier', SVC(kernel='linear', C=1))])
pipe_1.fit(X_train, y_train)

acc = cross_val_score(pipe_1, X_train, y_train, cv=2)
print(f'Training accuracy: {acc.mean() * 100: .2f} %')
```

```python
y_pred = pipe_1.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm).plot()
plt.title('Confusion Matrix')
```

```python
CR = classification_report(y_test, y_pred)
print(CR)
```

## Non-linear SVM for MNIST multiclass classification

```python
pipe_2 = Pipeline([('scaler', MinMaxScaler()), 
                   ('classifier', SVC(kernel='rbf', gamma=0.1, C=1))])
pipe_2.fit(X_train, y_train)

acc = cross_val_score(pipe_2, X_train, y_train, cv=2)
print(f'Training accuracy: {acc.mean() * 100: .2f} %')
```

```python
y_pred = pipe_2.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm).plot()
plt.title('Confusion Matrix')
```

```python
CR = classification_report(y_test, y_pred)
print(CR)
```
